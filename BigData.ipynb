```python
for item in re.findall(r'<td.*?</td?>', text, re.DOTALL):
    if ( re.findall(r'<td class="left">.*?</td>', item, re.DOTALL)):
        res = re.search(r'<a href="(.*?)>', item)
        if ( res != None):
            print("url", item)
        else:
            print("author", item)
```

```python
urls = []
titles = []
authors = []
prices = []
dates = []
brands = []

for item in re.findall(r'<td.*?</td?>', text, re.DOTALL):
    # url, title & author
    if ( re.findall(r'<td class="left">.*?</td>', item, re.DOTALL)):
        res = re.search(r'<a href="(.*?)>', item)
        if ( res != None):
            url = re.search(r'<a href="(.*?)>', item).group(1)
            url = "https://www.hanbit.co.kr" + url
            urls.append(url)
            title = re.sub(r'<.*?>', '', item)
            titles.append(title)
            print("url", url)
            print("title", title)
        else:
            author = re.sub(r'<.*?>', '', item)
            authors.append(author)
            print("author", author)
    # date
    if ( re.findall(r'<td>.*?</td>', item, re.DOTALL)):
        date = re.sub(r'<.*?>', '', item)
        dates.append(date)
        print("date", date)
    # price
    if ( re.findall(r'<td class="right">.*?</td>', item, re.DOTALL)):
        price = re.sub(r'<.*?>', '', item)
        prices.append(price)
        print("price", price)
    # brand
    if ( re.findall(r'<td class="brd_.*">.*?</td>', item, re.DOTALL)):
        brand = re.sub(r'<.*?>', '', item)
        brands.append(brand)
        print("brand", brand)
```

```python
import pandas as pd

df = pd.DataFrame (
    {
        "URL": urls,
        "TITLE": titles,
        "AUTHOR": authors,
        "DATE": dates,
        "PRICE": prices,
        "BRAND": brands
    }
)
```

```python
def no_space(text):
    text1 = re.sub('\nbsp: | &nbsp: | \n|\t|\r', '', text)
    text2 = re.sub('\n\n', '', text1)
    return text2
```

```python
def get_item(text, idx):
    text2 = ""
    cnt = 0
    for t in text:
        if( (len(text2) == 0) & (t == " ")):
            continue
        if( t == "\n"):
            continue
        if( cnt == idx-1):
            if(t != "|"):
                text2 = text2 + t
        if( t == '|'):
            cnt += 1
        if( cnt == idx):
            break
    return text2
```

```python
author_list = []
publisher_list = []
date_list = []
price_list = []
```

```python
import re
for b_detail in bestseller_list:
    temp = b_detail.find('div', {'class': 'author'})
    temp2 = temp.get_text()
    output = re.sub(r'저자', '', temp2)
    output = re.sub(r'더보기', '', output)
    output = no_space(output)
    author = get_item(output, 1)
    author_list.append(author)
    publisher = get_item(output, 2)
    publisher_list.append(publisher)
    date = get_item(output, 3)
    date_list.append(date)

    temp3 = b_detail.find('div', {'class': 'price'})
    price = temp3.find('strong').text
    price_list.append(price)
```

```python
def compute_error (y, y_pred, w_j):
    return (numpy.sum((w_j * numpy.not_equal(y, y_pred)),  axis =-1).astype(int))/numpy.sum(w_j)
```

```python
def compute_alpha(error):
    return 0.5*numpy.log(1-error + 1e-6)/(error+1e-6)
```

```python
def update_weights(w_j, alpha, y, y_pred):
    w_j_update = numpy.where(numpy.not_equal(y, y_pred),
                             w_j * numpy.exp(alpha * (numpy.not_equal(y, y_pred))),
                             w_j * numpy.exp(-alpha * (numpy.equal(y, y_pred))))
    w_j_update = w_j_update / numpy.sum(w_j_update)

    return w_j_update
```

```python
class AdaBoost:
    def __init__(self):
        self.alphas = []
        self.slearner = []
        self.M = None
        self.training_errors = []
        self.prediction_errors = []

    def fit (self, X, y, M = 150):
        self.M = M

        # w_j 초기화 -> w_j: tenser -> (1, 1, 1, 1, 1) / n
        w_j = numpy.array(numpy.ones(len(y))) / len(y)
        for m in range(0, M):
            # 1. weak learner를 훈련
            wlearner =  DecisionTreeClassifier(max_depth= 1)
            wlearner.fit (X, y, sample_weight = w_j)

            y_pred = wlearner.predict(X)

            self.slearner.append(wlearner)

            # 2. error 계산
            error = compute_error(y, y_pred, w_j)
            self.training_errors.append(error)

            # 3. alpha 계산
            alpha = compute_alpha(error)
            self.alphas.append(alpha)

            # 4. w_j를 업데이트
            w_j = update_weights(w_j, alpha, y, y_pred)

    def predict(self, X):
        weak_preds = numpy.zeros(shape=(X.shape[0], len(self.slearner)))
        for m in range(self.M):
            y_pred_m = self.slearner[m].predict(X).astype(float) * self.alphas[m]
            weak_preds[:,m] = y_pred_m
        y_pred = numpy.sign(numpy.sum(weak_preds, axis = -1))

        return y_pred    
```

```python
from sklearn.tree import DecisionTreeRegressor

class GradientBoost:
    def __init__(self, learning_rate, n_estimators, max_depth = 1):
        self.learning_rate = learning_rate
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.trees = []
    
    def fit(self, X, y):
        # 1. base classifier -> 평균
        self.f0 = y.mean()
        f_hat = self.f0

        for _ in range(self.n_estimators):
            # 1. residual 계산
            r = y - f_hat

            # 2. modify_clasifier 훈련
            tree = DecisionTreeRegressor(max_depth = self.max_depth, random_state = 11)
            tree.fit(X, r)
            
            # 3. hm
            h_m = tree.predict(X)

            # 4. f_hat에 추가
            f_hat += self.learning_rate * h_m
            
            self.trees.append(tree)

    def predict(self, X):
        f_hat = self.f0

        for m in range(self.n_estimators):
            f_hat += self.learning_rate * self.trees[m].predict(X)

        return f_hat
```

```python
import numpy as np

X = idata.data
y = idata.target
n = X.shape[0]

w = X[:, 0]
x = X[:, 1]
y = X[:, 2]
z = X[:, 3]

from matplotlib import pyplot as plt

plt.scatter(w, x, alpha=0.5)
plt.show()

plt.scatter(y, z, alpha=0.5)
plt.show()
```

```python
k = 3

centroid_w = np.random.uniform(min(w), max(w), k)
centroid_x = np.random.uniform(min(x), max(x), k)
centroid_y = np.random.uniform(min(y), max(y), k)
centroid_z = np.random.uniform(min(z), max(z), k)

centroid = list(zip(centroid_w, centroid_x, centroid_y, centroid_z))
centroid = np.array(centroid)
print(centroid)
```

```python
def distance(a, b):
    alen = len(a)
    sum = 0
    for aidx in range(alen):
        sum += (a[aidx] - b[aidx])**2
    return sum**0.5
```

```python
from copy import deepcopy

labels = np.zeros(n)
points = np.array(list(zip(w, x, y, z)))

centroid_old = np.zeros(centroid.shape)

error = np.zeros(k)
error_old = deepcopy (error)
```

```python
# error 초기화
for i in range(k):
    error[i] = distance(centroid_old[i], centroid[i])

idx = 0
while (error.any() != 0):
    # (2) 가장 가까운 centroid로 point를 clustering
    for i in range(n):
        distances = np.zeros(k)
        for j in range(k):
            distances[j] = distance(points[i], centroid[j])
        labels[i] = np.argmin(distances)
    # (3) 새로운 centroid를 계산
    centroid_old = deepcopy(centroid)
    for i in range(k):
        point = []
        for j in range(n):
            if ( labels[j] == i):
                point.append(points[j])
                
        centroid[i] = np.mean(point, axis = 0)

    for j in range(k):
        error[j] = distance(centroid_old[j], centroid[j])

    print(idx, "times over", error)
    idx += 1;
```
