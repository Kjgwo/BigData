```python
for item in re.findall(r'<td.*?</td?>', text, re.DOTALL):
    if ( re.findall(r'<td class="left">.*?</td>', item, re.DOTALL)):
        res = re.search(r'<a href="(.*?)>', item)
        if ( res != None):
            print("url", item)
        else:
            print("author", item)
```

```python
urls = []
titles = []
authors = []
prices = []
dates = []
brands = []

for item in re.findall(r'<td.*?</td?>', text, re.DOTALL):
    # url, title & author
    if ( re.findall(r'<td class="left">.*?</td>', item, re.DOTALL)):
        res = re.search(r'<a href="(.*?)>', item)
        if ( res != None):
            url = re.search(r'<a href="(.*?)>', item).group(1)
            url = "https://www.hanbit.co.kr" + url
            urls.append(url)
            title = re.sub(r'<.*?>', '', item)
            titles.append(title)
            print("url", url)
            print("title", title)
        else:
            author = re.sub(r'<.*?>', '', item)
            authors.append(author)
            print("author", author)
    # date
    if ( re.findall(r'<td>.*?</td>', item, re.DOTALL)):
        date = re.sub(r'<.*?>', '', item)
        dates.append(dates)
        print("date", date)
    # price
    if ( re.findall(r'<td class="right">.*?</td>', item, re.DOTALL)):
        price = re.sub(r'<.*?>', '', item)
        prices.append(price)
        print("price", price)
    # brand
    if ( re.findall(r'<td class="brd_.*">.*?</td>', item, re.DOTALL)):
        brand = re.sub(r'<.*?>', '', item)
        brands.append(brand)
        print("brand", brand)
```

```python
import pandas as pd

df = pd.DataFrame (
    {
        "URL": urls,
        "TITLE": titles,
        "AUTHOR": authors,
        "DATE": dates,
        "PRICE": prices,
        "BRAND": brands
    }
)
```
