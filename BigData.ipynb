```python
pd.concat([df1, df2])
```

```python
df1 = pd.DataFrame(
    {
        "key": ['k0', 'k1', 'k2', 'k3'],
     "A": ['a00', 'a01', 'a02', 'a03'],
     "B": ['a10', 'a11', 'a12', 'a13'],
    }
)
```

```python
df2 = pd.DataFrame(
    {
        "key": ['k1', 'k2', 'k3', 'k4'],
     "C": ['b00', 'b01', 'b02', 'b03'],
     "D": ['b10', 'b11', 'b12', 'b13'],
    }
)
```

```python
pd.merge(df1, df2, on = "key")
```
